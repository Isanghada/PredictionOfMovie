#### 감독 컬럼 정리

- 감독 데이터
  - 다수의 감독이 있는 경우 콤마(,)로 구분되어 있다.
  - 다수의 감독을 분리하여 1명의 대표감독으로 변경해준다.
    - ','를 기준으로 분리 오름차순 정렬 후 첫 번째 감독을 사용
    - `'안소니 루소,조 루소'`-`'조 루소, 안소니 루소'`와 같은 형태가 존재하기에 정렬을 진행한다.

```python
ranking_300['감독'].unique()
```

```python
# 다수의 감독 데이터 변경
for idx, row in ranking_300.iterrows():
    if ',' in row['감독']:
        directors = row['감독'].split(',')
        # 양 끝 공백 제거 후 정렬
        directors.sort(key=lambda x : x.strip())
        ranking_300.loc[idx, '감독'] = directors[0].strip()
```

#### 개봉일 개봉년도로 변경

## ◾배우 정보 추가

- 네이버 API를 이용하여 주연 배우 정보를 추가한다.

#### 네이버 API 함수

```python
import json
import urllib
```

```python
client_id = 'xzjv8j9BbPlDkjv3L4cP'
client_secret = 'qXSjpcpeln'
```

```python
def gen_url(movie, year):
    url = '''https://openapi.naver.com/v1/search/movie.json?query={query}&display={display}&start={start}&yearfrom={year}&yearto={year}'''
    encText = urllib.parse.quote(movie)

    return url.format(query=encText, display=100, start=1, year=year)
```

```python
def get_result_onpage(url):
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    return json.loads(response.read().decode("utf-8"))
```

```python
# <b> 태그를 없애는 함수
def delete_tag(input_str):
    input_str = input_str.replace('<b>', "")
    input_str = input_str.replace('</b>', "")
    return input_str
```

```python
import pandas as pd
def get_fields(json_data, movie, director):
    title = ''
    directors = []
    actors = ''
    for each in json_data['items']:
        try:    
            title = delete_tag(each['title']).strip()
            directors = each['director'].replace('|', '')
            actors = each['actor'].split('|')[:-1]
        except:
            continue
        if movie == title and director in directors:
            break
    return actors if actors else np.nan
```

#### 배우 정보 추출

```python
actor_list = []

for idx, row in ranking_300.iterrows():
    url = gen_url(str(row['영화명']), row['개봉년도'])
    json_result = get_result_onpage(url)
    result = get_fields(json_result, str(row['영화명']), row['감독'])
    actor_list.append(result)
    print('{}, {}---------------------------'.format(idx, row['영화명']))

len(actor_list)
```

#### 배우 정보 추가

```python
ranking_300['배우'] = actor_list
```
