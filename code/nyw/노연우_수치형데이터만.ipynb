```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
pd.set_option('display.max_columns',None)
```

```python
naver_data = pd.read_excel('./data/네이버_영화_데이터.xlsx', index_col=0)
naver_data.set_index("영화명",inplace=True)
naver_data.head(3)
```

```python
kobis_data = pd.read_excel('./data/전체_영화_데이터.xls')
kobis_data.head(3)
```

```python
kobis_data["주연배우"] = np.nan
kobis_data["평점"] = np.nan
kobis_data["평가자수"] = np.nan
kobis_data["상영시간"] = np.nan
```

```python
kobis_data.head()
```

```python
kobis_data['영화명'] = kobis_data['영화명'].astype('str')
kobis_data.set_index('영화명', inplace=True)
```

```python
for i in range(len(naver_data)):#len(naver_data)):
    movieName = naver_data.index[i]
    kobis_data.loc[movieName,'주연배우'] = naver_data.loc[str(movieName),'주연배우']
    kobis_data.loc[movieName,'평점'] = naver_data.loc[str(movieName),'평점']
    kobis_data.loc[movieName,'평가자수'] = naver_data.loc[str(movieName),'평가자수']
    kobis_data.loc[movieName,'상영시간'] = naver_data.loc[str(movieName),'상영시간']
```

```python
df_sum = kobis_data[(kobis_data['주연배우'].notnull()) & (kobis_data['전국스크린수']>50) & (kobis_data['전국관객수']>1000)]
```

```python
df_sum = df_sum.reset_index()
```

```python
df_sum
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df_1119 = df_sum[(df_sum['개봉일'] <'2020-01-01') & (df_sum['개봉일'] >'2010-12-31')]
df_20 = df_sum[(df_sum['개봉일'] >'2019-12-31')]
```

```python
X = df_1119[['전국스크린수','전국매출액', '상영시간','평점','평가자수']] # ,'평가자수'
y = df_1119['전국관객수']
```

```python
X.corr()
```

```python
len(X), len(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
params = {
    'max_depth' : [4, 6, 8, 10],           
    'n_estimators' : [50, 100, 200, 500],   
    'min_samples_leaf' : [5, 12],      
    'min_samples_split' : [5, 12]      
}
# n_jobs : cpu 코어의 개수
rfr_clf = RandomForestRegressor(n_jobs=-1)
grid_cv = GridSearchCV(rfr_clf, param_grid=params, cv=5, n_jobs=-1)
grid_cv.fit(X_train, y_train)
```

```python
cv_results_df = pd.DataFrame(grid_cv.cv_results_)
cv_results_df.columns
```

```python
target_col = ['rank_test_score', 'mean_test_score', 'param_n_estimators', 'param_max_depth']
cv_results_df[target_col].sort_values(by='rank_test_score').head()
```

```python
grid_cv.best_params_
```

```python
best_rfr_clf = grid_cv.best_estimator_

pred_tr = best_rfr_clf.predict(X_train)
pred_test = best_rfr_clf.predict(X_test)
rmse_tr = (np.sqrt(mean_squared_error(y_train, pred_tr)))
rmse_test = (np.sqrt(mean_squared_error(y_test, pred_test)))

print('RMSE of Train Data : {}'.format(rmse_tr))
print('RMSE of Test Data : {}'.format(rmse_test))
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8, 8))
plt.scatter(y_test, pred_test)
plt.title('Real vs Predicted')
plt.show()
```
