```python
import pandas as pd
import numpy as np
```

```python
naver_data = pd.read_excel('./data/네이버_영화_데이터.xlsx', index_col=0)
naver_data.set_index("영화명",inplace=True)
naver_data.head(3)
```

```python
kobis_data = pd.read_excel('./data/전체_영화_데이터.xls')
kobis_data.head(3)
```

```python
kobis_data["주연배우"] = np.nan
kobis_data["평점"] = np.nan
kobis_data["평가자수"] = np.nan
kobis_data["상영시간"] = np.nan
```

```python
kobis_data.head()
```

```python
kobis_data.loc[33,'주연배우'], naver_data.loc['33','주연배우']
```

```python
for i in range(10):#len(naver_data)):
    movieName = naver_data.index[i]
    kobis_data.loc[movieName,'주연배우'] = naver_data.loc[str(movieName),'주연배우']
    kobis_data.loc[movieName,'평점'] = naver_data.loc[str(movieName),'평점']
    kobis_data.loc[movieName,'평가자수'] = naver_data.loc[str(movieName),'평가자수']
    kobis_data.loc[movieName,'상영시간'] = naver_data.loc[str(movieName),'상영시간']
```
