```python
kobis['영화명'] = kobis['영화명'].astype('str')
```

```python
naver_kobis = pd.merge(kobis, naver, left_on=('영화명','감독'), right_on=('영화명','감독'), how='left')
naver_kobis.head()
```

```python
all = naver_kobis[naver_kobis['주연배우'].isnull() == False]
```

```python
all = all.reset_index()
```

```python
all.drop(['index','순번'], axis =1, inplace=True)
all
```
